Эта грамматика задает язык $$\{a^{m + 3n}cb^{4m + 2n}| m,n \in N_{\ge 0}\}$$
понятно что для каждой строки можно сначала применить все первые правила, 
а потом все вторые (а 3 применяется 1 раз в конце).
Новая грамматика: 
$$S -> aS_1bbbb \\
   S_1 -> aS_1bbbb | S_2 \\
   S_2 -> aaaS_2bb | c \\
$$
понятно что она однозначна, т.к мы фиксировали порядок применения правил и разным парам {m, n} соответсвуют разные слова

# 3

Это язык ($A$) над алфавитом {$a, b$} в котором $|a| = 2\cdot |b|$
и в любом префиксе $\ge 2\cdot |b|$

Понятно, что язык, порождаемый этой грамматикой лежит в описанном выше (индукция по длинне слова).

Нужно проверить обратное включение.

Пусть $w \in A$, возьмем минимальный префикс, который принадлежит языку (в любом префиксе $|a| \ge 2\cdot |b|$ , разница между $a$ и $2b$ хотя бы в конце будет 0) т.е $w = st \{s \in A\}$

Понятно, что $t \in A$ (разница между $a$ и $b$ в его префиксе соответсвует разнице между ними в соотв префиксе $w$)
Значит осталось проверить что s может быть получена через $aFaFb$

$s$ точно начинается с $a$ и точно заканчивается $b$, т.е $s = atb$, у $t$ $|a| = 2\cdot |b| + 1$

Т.к на прошлом шаге префикс был минимальным, то $|a| - 2\cdot |b| \ge 0$ на любом префиксе $t$. 

Понятно что если существует разложение $t = a_1aa_2, a_1, a_2 \in A$, то оно единственно, иначе нарушится уравнение на $|a|$ и $|b|$. 

Оно существует потому что изначально $|a| - 2\cdot |b| = 0$ и в конце равна 1 и увеличивается за раз на 0 или 1, т.е точно есть момент перехода от 0 к 1.

Значит $t$ может быть получена как $FaF$, тогда $s$ может быть получена как $aFaFb$

Но тогда $w$ может быть получено как $aFaFbF$
